```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Assuming your new data is in a CSV file named 'combined_data.csv'
file_path_combined = r'D:\Me\concordia\Notes\SE4AI\project\Implementation\Implementation-Git\SOEN-691-Project\data\sorted_weighted_average_by_ML_Pipeline_Type.csv'


# Read the new data into a DataFrame
df_combined = pd.read_csv(file_path_combined)

# Create a grouped bar chart
plt.figure(figsize=(12, 8))
sns.barplot(x='ML Pipeline Stage', y='Weighted Average', hue='ML TD Type', data=df_combined, palette='viridis')

# Add labels and title
plt.xlabel('ML Pipeline Stage')
plt.ylabel('Weighted Average')
plt.title('Weighted Average for ML TD Type in Each ML Pipeline Stage')

# Display the legend
plt.legend(title='ML TD Type', bbox_to_anchor=(1, 1))

# Rotate x-axis labels for better readability
plt.xticks(rotation=45)

# Display the plot
plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Assuming your new data is in a CSV file named 'combined_data.csv'
file_path_combined = r'D:\Me\concordia\Notes\SE4AI\project\Implementation\Implementation-Git\SOEN-691-Project\data\sorted_weighted_average_by_ML_Pipeline_Type.csv'

# Read the new data into a DataFrame
df_combined = pd.read_csv(file_path_combined)

# Define a distinct color palette
distinct_palette = sns.color_palette("husl", n_colors=len(df_combined['ML TD Type'].unique()))

# Create a grouped bar chart with distinct colors
plt.figure(figsize=(12, 8))
sns.barplot(x='ML Pipeline Stage', y='Weighted Average', hue='ML TD Type', data=df_combined, palette=distinct_palette)

# Add labels and title
plt.xlabel('ML Pipeline Stage')
plt.ylabel('Weighted Average')
plt.title('Weighted Average for ML TD Type in Each ML Pipeline Stage')

# Display the legend
plt.legend(title='ML TD Type', bbox_to_anchor=(1, 1))

# Rotate x-axis labels for better readability
plt.xticks(rotation=45)

# Display the plot
plt.show()
```
